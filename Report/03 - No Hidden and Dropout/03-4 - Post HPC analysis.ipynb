# Analysis of the results from HPC

In this section of the report we experimented with two different forms of a model, and will see a similar model in section 05 as we were looking for consistent implementation. In this section we first tried a model with no hidden layers, only using an input and output layer.

The next model we implemented used dropout layers. These are supposed to prevent overfitting by dropping a percentage of the input data. This was implemented using a GridSearch method adapted from the work in section 05 where it will be explained in better detail.

We import packages as usual:

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from urllib.request import urlopen
```

## No hidden layers model

We import our epochs, training data accuracy, and test accuracy outputs from the HPC that we have copied back over to my home PC and pushed to GitHub.

```python
epochs1 = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/epochs_nohidden.pkl'))
trainacc1 = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/trainacc_nohidden.pkl'))
testacc1 = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/testacc_nohidden.pkl'))
```

We will plot these values to see how the accuracy is affected with different numbers of epochs.

```python
plt.plot(epochs1, trainacc1, color = 'r',label="Training Accuracy")
plt.plot(epochs1, testacc1, color='b',label="Testing Accuracy")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('A graph showing accuracy on training and test data over 1000 epochs of training')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, matthews_corrcoef, cohen_kappa_score, accuracy_score, average_precision_score, roc_auc_score
import seaborn as sn
```

We predicted our Y values using the `modeldump.py` file tuned for each model loading our saved model on the HPC. You can find this in the Appendix folder. We then load the saved pickled files along with our correct Y test values to compare with.

```python
Y_pred = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/Y_predictions.pkl'))
Y_test_c = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/Y_test_c2.pkl'))
```

We plot a visualisation of our confusion matrix.

```python
a = confusion_matrix(Y_test_c, Y_pred)
```

```python
plt.figure(figsize = (10,7))
sn.heatmap(a, annot=True)
```

## Dropout Layer Model

We repeat the above process for our model featuring dropout layers.

```python
epochs2 = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/epochs_dropout.pkl'))
trainacc2 = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/trainacc_dropout.pkl'))
testacc2 = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/testacc_dropout.pkl'))
```

```python
plt.plot(epochs2, trainacc2, color = 'r',label="Training Accuracy")
plt.plot(epochs2, testacc2, color='b',label="Testing Accuracy")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('A graph showing accuracy on training and test data over 1000 epochs of training')
plt.legend()
plt.show()
```

```python
Y_pred = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/Y_predictions_drop.pkl'))
Y_test_c = pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-04/raw/main/Gabriel%20Grant/HPC/Y_test_c_drop.pkl'))
```

```python
a = confusion_matrix(Y_test_c, Y_pred)
```

```python
plt.figure(figsize = (10,7))
sn.heatmap(a, annot=True)
```

As is epected the training accuracy is higher than the testing accuracy, but they follow the same trend. Also predictably there is more variation in the accuracy on the testing data than on the training data.

We see that there is not much difference in the overall accuracy between the model without dropout layers and the model with no hidden layers, though no hidden model does seem to have a slightly higher average. However, it seems the dropout model converges considerably quicker than the no hidden model, but the no hidden model has considerably more consistent results; there is a lot less jitter in the line on the graph. This is particularly apparent in the testing accuracy, as the accuracy of the dropout model fluctuates quite considerably in comparison to the no hidden model.
